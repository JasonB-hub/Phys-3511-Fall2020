```python
#Problem 1
for i in range (3,20):
    f3=(math.log(math.factorial(i)))
    f4=(i*math.log(i)-(i))
    print(f3)
    print(f4)
    print(" ")
    plot.plot(f3,f4,'r')
    plot.xlabel('Exact Value of ln(n!)')
    plot.ylabel('Stirlings Approximation of ln(n!)')
plot.show()
```

```python
#Problem 1
exact = math.log(math.factorial(3))
approx = (3*math.log(3)-3)
error = ((exact-approx)/exact)
print("%.3f" % error)
```
